```python
def aggregate_metrics(df):
    # Dictionary to store columns to be aggregated
    metric_groups = {}
    standalone_cols = []
    
    # Find columns to be aggregated (those ending with numbers)
    for col in df.columns:
        base_name = ''.join([c for c in col if not c.isdigit()])
        if base_name != col:  # If the column name contained numbers
            if base_name not in metric_groups:
                metric_groups[base_name] = []
            metric_groups[base_name].append(col)
        else:
            standalone_cols.append(col)
    
    # Create new dataframe with aggregated metrics
    aggregated_data = pd.DataFrame()
    
    # Add standalone columns
    for col in standalone_cols:
        aggregated_data[col] = df[col]
    
    # Calculate mean for grouped metrics
    for base_name, columns in metric_groups.items():
        aggregated_data[base_name] = df[columns].astype(float).mean(axis=1)
    
    return aggregated_data

# Apply aggregation to POST_short and PRE_short
POST_final = aggregate_metrics(POST_short)
PRE_final = aggregate_metrics(PRE_short)

# If POST_interaction exists, aggregate it as well
POST_interaction_final = aggregate_metrics(POST_interaction)

print("\nPRE_final columns:", PRE_final.columns.tolist())
print("\nPOST_final columns:", POST_final.columns.tolist())
print("\nPOST_interaction_final columns:", POST_interaction_final.columns.tolist())

```

```python
import pandas as pd
from scipy import stats

# --- 1) Compute the delta (post - pre) for each metric ---
metrics = [
    'anxious', 'overwhelmed', 'relaxed', 'energetic', 'motivated',
    'AHS', 'FSQ_Similarity', 'FSQ_Vividness', 'FSQ_Positivity',
    'Future_self_continuity'
]
df = merged_df.copy()
# assuming your dataframe is called df
for m in metrics:
    # In the provided code snippet, `df` is being used as a placeholder variable to represent a pandas DataFrame where the delta values for each metric (post - pre) are being computed and stored. The actual DataFrame object is not explicitly defined in the code snippet provided, so it is assumed that `df` represents the DataFrame where these calculations are being performed.
    df[f'{m}_delta'] = merged_df[f'{m}_post'] - merged_df[f'{m}_pre']

# --- 2) Run a one-way ANOVA for each metric delta ---
anova_results = []
for m in metrics:
    # group the delta values by condition
    groups = [
        df.loc[df['cond_post'] == cond, f'{m}_delta'].dropna()
        for cond in ['control', 'text', 'avatar', 'voice']
    ]
    F, p = stats.f_oneway(*groups)
    anova_results.append({'Metric': m, 'F-statistic': F, 'p-value': p})

anova_df = pd.DataFrame(anova_results)
anova_df['Significance'] = anova_df['p-value'].apply(
    lambda p: '***' if p < 0.001 else '**' if p < 0.01 else '*' if p < 0.05 else ''
)

print("\nANOVA Results (Î” = Post - Pre):")
print(anova_df.to_string(index=False))

# ------------------------------------------------------------------
# 4) Post-hoc Tukey HSD for significant metrics
# ------------------------------------------------------------------
from statsmodels.stats.multicomp import pairwise_tukeyhsd

print("\n=== Tukey HSD post-hoc tests for significant metrics ===")
for _, row in anova_df.iterrows():
    if row['Significance'] != '':
        m = row['Metric']
        metric_name = f'{m}_delta'
        data = df[[metric_name, 'cond_post']].dropna()

        if data.empty:
            continue

        tukey = pairwise_tukeyhsd(
            endog=data[metric_name],
            groups=data['cond_post'],
            alpha=0.05
        )
        print(f"\n--- {m.upper()} ---")
        print(tukey.summary())

```

```python

# Create a function to calculate Cohen's d
def cohens_d(group1, group2):
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_se = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    return (np.mean(group1) - np.mean(group2)) / pooled_se

# Create a function to interpret Cohen's d
def interpret_cohens_d(d):
    if abs(d) < 0.2:
        return "Negligible"
    elif abs(d) < 0.5:
        return "Small"
    elif abs(d) < 0.8:
        return "Medium"
    else:
        return "Large"

# Initialize lists to store results
comparison_results = []

# For each significant metric
for metric in significant_metrics:
    # Get the differences for each condition
    for cond1 in ['avatar', 'text', 'voice']:
        for cond2 in ['control']:
            # Skip if comparing condition with itself
            if cond1 == cond2:
                continue
                
            # Get the pre and post values
            pre_metric = metric
            post_metric = metric if metric in POST_final.columns else metric[0].upper() + metric[1:]
            
            # Calculate differences for both conditions
            cond1_mask = merged_df['cond_pre'] == cond1
            cond2_mask = merged_df['cond_pre'] == cond2
            
            diff1 = merged_df[cond1_mask][f'{metric}_post'] - merged_df[cond1_mask][f'{metric}_pre']
            diff2 = merged_df[cond2_mask][f'{metric}_post'] - merged_df[cond2_mask][f'{metric}_pre']
            
            # Calculate mean difference
            mean_diff = np.mean(diff1) - np.mean(diff2)
            
            # Calculate Cohen's d
            d = cohens_d(diff1, diff2)
            effect_size = interpret_cohens_d(d)
            
            # Perform t-test
            t_stat, p_val = stats.ttest_ind(diff1, diff2)
            
            # Add significance stars
            stars = ''
            if p_val < 0.001:
                stars = '***'
            elif p_val < 0.01:
                stars = '**'
            elif p_val < 0.05:
                stars = '*'
            
            # Add to results
            comparison_results.append({
                'Measure': metric,
                'Comparison': f'{cond1} vs {cond2}',
                'Mean Diff.': f"{mean_diff:.2f}",
                "Cohen's d": f"{d:.2f}",
                'Effect Size': effect_size,
                'p-value': f"{p_val:.3f}{stars}"
            })

# Create DataFrame and display results
comparison_df = pd.DataFrame(comparison_results)
print("\nDetailed Comparison Results:")
print(comparison_df.to_string(index=False))

# Save to CSV (relative path)
results_folder = os.path.join(os.getcwd(), 'Results')
os.makedirs(results_folder, exist_ok=True)
comparison_path = os.path.join(results_folder, 'detailed_comparisons.csv')
comparison_df.to_csv(comparison_path, index=False)
print(f"\nDetailed comparison results saved to: {comparison_path}")


```

```python
from statsmodels.stats.multitest import multipletests  # NEW: for FDR correction

# Define metrics and conditions - ADDED FSQ_Similarity
fsq_metrics = ['FSQ_Vividness', 'FSQ_Positivity', 'FSQ_Similarity', 'Future_self_continuity']
conditions = ['avatar', 'text', 'voice', 'control']
colors = {
    'avatar':  '#9B4F96',
    'text':    '#C17FBE',
    'voice':   '#DFB6DD',
    'control': '#808080'
}

# Helper to map p-values to significance stars
def p_to_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    else:
        return ''

# Create figure with 4 subplots
fig, axes = plt.subplots(1, 4, figsize=(24, 6))

for idx, metric in enumerate(fsq_metrics):
    ax = axes[idx]
    x = np.arange(len(conditions))
    width = 0.6
    
    # -------------------------------------------------------------------------
    # 1) Compute adjusted means (POST controlling for PRE)
    # -------------------------------------------------------------------------
    means = []
    sems = []
    
    # grand mean of PRE for this metric (across all participants)
    grand_mean_pre = merged_df[f'{metric}_pre'].mean()
    
    for cond in conditions:
        cond_data = merged_df[merged_df['cond_post'] == cond]
        pre_vals = cond_data[f'{metric}_pre'].values
        post_vals = cond_data[f'{metric}_post'].values
        
        if len(pre_vals) > 0:
            cond_mean_pre = pre_vals.mean()
            adjustment = cond_mean_pre - grand_mean_pre
            adjusted_mean = post_vals.mean() - adjustment
            
            means.append(adjusted_mean)
            sems.append(post_vals.std(ddof=1) / np.sqrt(len(post_vals)))
        else:
            means.append(np.nan)
            sems.append(np.nan)
    
    # Plot bars
    bars = ax.bar(
        x, means, width, yerr=sems,
        color=[colors[c] for c in conditions],
        error_kw=dict(lw=1.5, capthick=1.5, capsize=8, ecolor='black'),
        edgecolor='black',
        linewidth=1.2,
        alpha=0.85
    )
    
    # -------------------------------------------------------------------------
    # 2) Compute dynamic significance vs control (change scores + FDR-BH)
    # -------------------------------------------------------------------------
    finite_means = [m for m in means if np.isfinite(m)]
    finite_sems = [s for s in sems if np.isfinite(s)]
    
    if len(finite_means) > 0:
        data_max = max(finite_means)
        max_sem = max(finite_sems) if finite_sems else 0
        base_height = data_max + max_sem
        yrange = (max(finite_means) - min(finite_means)) + 2 * max_sem + 1e-8
        spacing = 0.12 * yrange
        tick_h = 0.03 * yrange
        
        # helper: change scores (post - pre) for a condition
        def diffs_for(cond_name):
            sub = merged_df[merged_df['cond_post'] == cond_name]
            pre = sub[f'{metric}_pre'].values
            post = sub[f'{metric}_post'].values
            n = min(len(pre), len(post))
            if n <= 1:
                return np.array([])
            return post[:n] - pre[:n]
        
        control_diffs = diffs_for('control')
        
        comparisons = []   # [(cond_name, raw_p), ...]
        if len(control_diffs) > 1:
            for cond in conditions:
                if cond == 'control':
                    continue
                d1 = diffs_for(cond)
                if len(d1) > 1:
                    _, p = stats.ttest_ind(d1, control_diffs,
                                           equal_var=False, nan_policy='omit')
                    if np.isfinite(p):
                        comparisons.append((cond, p))
        
        # FDR-correct across all comparisons for this metric
        if comparisons:
            cond_names = [c for (c, _) in comparisons]
            pvals = np.array([p for (_, p) in comparisons])
            
            reject, pvals_corr, _, _ = multipletests(pvals, alpha=0.05,
                                                     method='fdr_bh')
            
            # Place brackets only for significant comparisons after FDR
            level = 1
            for cond_name, p_corr, rej in zip(cond_names, pvals_corr, reject):
                if not rej:
                    continue  # skip non-significant after FDR
                
                stars = p_to_stars(p_corr)
                if stars == '':
                    continue  # no visible annotation
                
                i1 = conditions.index(cond_name)
                i2 = conditions.index('control')
                
                line_y = base_height + spacing * level
                
                # Draw bracket
                ax.plot(
                    [x[i1], x[i1], x[i2], x[i2]],
                    [line_y - tick_h, line_y, line_y, line_y - tick_h],
                    color='black', linewidth=1.2
                )
                
                # Add stars
                ax.text(
                    (x[i1] + x[i2]) / 2,
                    line_y + tick_h * 0.8,
                    stars,
                    ha='center', va='bottom',
                    fontsize=12, fontweight='bold'
                )
                
                level += 1
            
            # Set y-limits with some headroom
            y_min = min(finite_means) - max_sem - 0.3
            # use the last level used; if no sigs, level is still 1
            y_max = base_height + spacing * max(level, 2)
            ax.set_ylim(y_min, y_max)
    
    # -------------------------------------------------------------------------
    # 3) Cosmetics
    # -------------------------------------------------------------------------
    ax.set_ylabel('Adjusted Mean Score', fontsize=12, fontweight='bold')
    ax.set_title(metric.replace('_', ' '), fontsize=14, fontweight='bold', pad=15)
    ax.set_xticks(x)
    ax.set_xticklabels([c.capitalize() for c in conditions],
                       fontsize=11, fontweight='bold')
    ax.grid(True, axis='y', alpha=0.3, linestyle='--', linewidth=0.8)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(1.2)
    ax.spines['bottom'].set_linewidth(1.2)
    ax.set_facecolor('#FAFAFA')

# Add overall title with caption
fig.suptitle('Adjusted Post-Intervention Future-Self Metrics by Condition',
             fontsize=16, fontweight='bold', y=0.98)

plt.tight_layout(rect=[0, 0.05, 1, 0.95])

# Add caption at bottom
fig.text(
    0.5, 0.01,
    'Adjusted means control for baseline scores. Error bars = SEM. '
    'Significance brackets show FDR-corrected comparisons vs. control '
    '(based on change scores: POSTâ€“PRE; *q<0.05, **q<0.01, ***q<0.001).',
    ha='center', fontsize=10, style='italic', wrap=True
)


plt.show()
print(f"âœ… Figure saved to: {save_path}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import os

# =============================================================================
# AUTOMATED PLOTTING WITHOUT FDR CORRECTION
# =============================================================================

# Define colors and metrics
colors = {
    'avatar': '#9B4F96',  # Dark purple
    'text':   '#C17FBE',  # Medium purple
    'voice':  '#DFB6DD',  # Light purple
    'control':'#808080'   # Grey
}

metrics = ['UES', 'Realism', 'Trust', 'Persuasion']
conditions = ['avatar', 'text', 'voice', 'control']

# =============================================================================
# COMPUTE MEANS AND SDs FROM ACTUAL DATA
# =============================================================================
means = {}
stds = {}
ses = {}  # Standard errors for error bars

for metric in metrics:
    metric_means = []
    metric_stds = []
    metric_ses = []
    
    for cond in conditions:
        vals = POST_interaction_final[POST_interaction_final['cond'] == cond][metric]
        vals = vals.dropna()  # Remove NaN values
        
        metric_means.append(vals.mean())
        metric_stds.append(vals.std(ddof=1))
        metric_ses.append(vals.std(ddof=1) / np.sqrt(len(vals)))
    
    means[metric] = metric_means
    stds[metric] = metric_stds
    ses[metric] = metric_ses

print("="*80)
print("DESCRIPTIVE STATISTICS")
print("="*80)
for metric in metrics:
    print(f"\n{metric}:")
    for i, cond in enumerate(conditions):
        print(f"  {cond.capitalize()}: M={means[metric][i]:.2f}, SD={stds[metric][i]:.2f}")

# =============================================================================
# PERFORM ANOVA AND TUKEY POST-HOC TESTS
# =============================================================================
print("\n" + "="*80)
print("ANOVA RESULTS")
print("="*80)

anova_results = []
sig_pairs = {}

for metric in metrics:
    # Get data for each condition
    groups = []
    for cond in conditions:
        vals = POST_interaction_final[POST_interaction_final['cond'] == cond][metric].dropna()
        groups.append(vals)
    
    # Perform one-way ANOVA
    f_stat, p_val = stats.f_oneway(*groups)
    
    # Determine significance
    if p_val < 0.001:
        stars = '***'
    elif p_val < 0.01:
        stars = '**'
    elif p_val < 0.05:
        stars = '*'
    else:
        stars = 'n.s.'
    
    anova_results.append({
        'Metric': metric,
        'F-statistic': f_stat,
        'p-value': p_val,
        'Significance': stars
    })
    
    print(f"\n{metric}: F({len(conditions)-1}, {sum(len(g) for g in groups)-len(conditions)}) = {f_stat:.2f}, p = {p_val:.4f} {stars}")
    
    # If significant, perform Tukey's HSD post-hoc test
    if p_val < 0.05:
        # Prepare data for Tukey test
        values = []
        group_labels = []
        for cond in conditions:
            vals = POST_interaction_final[POST_interaction_final['cond'] == cond][metric].dropna()
            values.extend(vals)
            group_labels.extend([cond] * len(vals))
        
        # Perform Tukey's HSD
        tukey = pairwise_tukeyhsd(values, group_labels, alpha=0.05)
        
        print(f"\nTukey's HSD Post-hoc Test for {metric}:")
        print(tukey)
        
        # Extract significant pairs from Tukey results
        metric_sig_pairs = []
        tukey_df = pd.DataFrame(data=tukey.summary().data[1:], columns=tukey.summary().data[0])
        
        for _, row in tukey_df.iterrows():
            if row['reject']:  # If null hypothesis is rejected
                group1 = row['group1']
                group2 = row['group2']
                p_adj = float(row['p-adj'])
                
                # Determine significance stars
                if p_adj < 0.001:
                    sig_stars = '***'
                elif p_adj < 0.01:
                    sig_stars = '**'
                elif p_adj < 0.05:
                    sig_stars = '*'
                else:
                    continue
                
                metric_sig_pairs.append(((group1.capitalize(), group2.capitalize()), sig_stars, p_adj))
        
        if metric_sig_pairs:
            sig_pairs[metric] = metric_sig_pairs

# Display ANOVA results
anova_df = pd.DataFrame(anova_results)
print("\n" + "="*80)
print("ANOVA SUMMARY TABLE")
print("="*80)
print(anova_df.to_string(index=False))

print("\n" + "="*80)
print("SIGNIFICANT PAIRWISE COMPARISONS (Tukey's HSD)")
print("="*80)
if sig_pairs:
    for metric, pairs in sig_pairs.items():
        print(f"\n{metric}:")
        for (cond1, cond2), stars, p_adj in pairs:
            print(f"  {cond1} vs {cond2}: {stars} (p-adj = {p_adj:.4f})")
else:
    print("No significant pairwise differences found")

# =============================================================================
# CREATE PLOT WITH STANDARD DEVIATIONS - ENHANCED DESIGN
# =============================================================================
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
axes = axes.flatten()

conditions_display = ['Avatar', 'Text', 'Voice', 'Control']
x = np.arange(len(conditions_display))
width = 0.65

for idx, metric in enumerate(metrics):
    ax = axes[idx]
    
    # Bars with SD error bars - ENHANCED STYLING
    bars = ax.bar(
        x, means[metric], width, yerr=stds[metric],
        color=[colors[c.lower()] for c in conditions_display],
        error_kw=dict(lw=1.5, capthick=1.5, capsize=8, ecolor='black'),
        edgecolor='black',
        linewidth=1.2,
        alpha=0.85
    )
    
    # Base formatting - ENHANCED
    ax.set_ylabel('Score', fontsize=12, fontweight='bold')
    ax.set_title(metric, fontsize=14, fontweight='bold', pad=15)
    ax.set_xticks(x)
    ax.set_xticklabels(conditions_display, fontsize=11, fontweight='bold')
    ax.grid(True, axis='y', alpha=0.3, linestyle='--', linewidth=0.8)
    
    # Clean up spines - NEW
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(1.2)
    ax.spines['bottom'].set_linewidth(1.2)
    ax.set_facecolor('#FAFAFA')
    
    # Calculate y-axis limits dynamically
    bar_tops = np.array(means[metric]) + np.array(stds[metric])
    finite_means = [m for m in means[metric] if np.isfinite(m)]
    finite_stds = [s for s in stds[metric] if np.isfinite(s) and s > 0]
    
    if len(finite_means) > 0:
        data_min = min(finite_means)
        data_max = max(finite_means)
        max_err = max(finite_stds) if len(finite_stds) > 0 else 0
        max_bar_top = max(bar_tops)
        
        # Dynamic spacing calculation
        base_height = max_bar_top + max_err * 1.5
        yrange = abs(data_max - data_min) + max_err * 2 + 0.5
        spacing = 0.12 * yrange
        ytick = 0.03 * yrange
        
        # Add significance lines if they exist for this metric
        if metric in sig_pairs:
            level = 1
            
            for (cond1, cond2), sig_level, p_adj in sig_pairs[metric]:
                try:
                    idx1 = conditions_display.index(cond1)
                    idx2 = conditions_display.index(cond2)
                except ValueError:
                    print(f"Warning: Could not find condition {cond1} or {cond2}")
                    continue
                
                # Calculate y position for this significance line
                y_pos = base_height + spacing * level
                
                # Draw the line with vertical ticks - ENHANCED STYLE
                x1, x2 = x[idx1], x[idx2]
                ax.plot([x1, x1, x2, x2], 
                       [y_pos - ytick, y_pos, y_pos, y_pos - ytick],
                       color='black', linewidth=1.2)
                
                # Add significance stars - ENHANCED
                ax.text((x1 + x2) / 2, y_pos + ytick * 0.8, sig_level, 
                       ha='center', va='bottom', fontsize=12, fontweight='bold')
                
                level += 1
            
            # Set y-limits with room for significance bars
            y_min = data_min - max_err * 2
            y_max = base_height + spacing * (level + 0.5)
            ax.set_ylim(y_min, y_max)
        else:
            # No significance bars, just set reasonable limits
            y_min = 0
            y_max = max(7.0, max_bar_top * 1.2)
            ax.set_ylim(y_min, y_max)
    else:
        ax.set_ylim(0, 7)

# Add overall title - NEW
fig.suptitle('Post-Intervention Interaction Quality Metrics by Condition\n(One-way ANOVA with Tukey\'s HSD post-hoc tests)',
             fontsize=16, fontweight='bold', y=0.995)

plt.tight_layout(rect=[0, 0, 1, 0.98])

# Save figure
graphs_folder = os.path.join(os.getcwd(), 'Results', 'Graphs')
os.makedirs(graphs_folder, exist_ok=True)
save_path = os.path.join(graphs_folder, 'interaction_metrics_comparison.png')
fig.savefig(save_path, dpi=300, bbox_inches='tight', facecolor='white')
print(f"\nâœ… Figure saved to: {save_path}")

plt.show()

# =============================================================================
# CREATE DETAILED COMPARISON TABLE (VS CONTROL)
# =============================================================================
print("\n" + "="*80)
print("DETAILED COMPARISONS VS CONTROL - not corrected for multiple comparisons")
print("="*80)

def cohens_d(group1, group2):
    """Calculate Cohen's d effect size"""
    n1, n2 = len(group1), len(group2)
    var1, var2 = np.var(group1, ddof=1), np.var(group2, ddof=1)
    pooled_std = np.sqrt(((n1 - 1) * var1 + (n2 - 1) * var2) / (n1 + n2 - 2))
    return (np.mean(group1) - np.mean(group2)) / pooled_std

def interpret_cohens_d(d):
    """Interpret Cohen's d magnitude"""
    abs_d = abs(d)
    if abs_d < 0.2:
        return "Negligible"
    elif abs_d < 0.5:
        return "Small"
    elif abs_d < 0.8:
        return "Medium"
    else:
        return "Large"

comparison_rows = []

for metric in metrics:
    control_vals = POST_interaction_final[POST_interaction_final['cond'] == 'control'][metric].dropna()
    
    for cond in ['avatar', 'text', 'voice']:
        cond_vals = POST_interaction_final[POST_interaction_final['cond'] == cond][metric].dropna()
        
        # Calculate statistics
        mean_diff = cond_vals.mean() - control_vals.mean()
        d = cohens_d(cond_vals, control_vals)
        effect_size = interpret_cohens_d(d)
        
        # Perform t-test
        t_stat, p_val = stats.ttest_ind(cond_vals, control_vals, equal_var=False)
        
        # Significance stars
        if p_val < 0.001:
            stars = '***'
        elif p_val < 0.01:
            stars = '**'
        elif p_val < 0.05:
            stars = '*'
        else:
            stars = ''
        
        comparison_rows.append({
            'Metric': metric,
            'Comparison': f'{cond.capitalize()} vs Control',
            'Mean Diff': f"{mean_diff:.2f}",
            "Cohen's d": f"{d:.2f}",
            'Effect Size': effect_size,
            't-statistic': f"{t_stat:.2f}",
            'p-value': f"{p_val:.4f}",
            'Sig': stars
        })

comparison_df = pd.DataFrame(comparison_rows)
print(comparison_df.to_string(index=False))

# Save comparison table
comparison_path = '/Users/ge93mey/Desktop/MIT_Media_Lab/Cyborg Psychology/IUI_Analysis/Results/Graphs/interaction_comparisons_vs_control.csv'
os.makedirs(os.path.dirname(comparison_path), exist_ok=True)
comparison_df.to_csv(comparison_path, index=False)
print(f"âœ… Comparison table saved to: {comparison_path}")

# =============================================================================
# SAVE SUMMARY TABLE
# =============================================================================
summary_rows = []
for metric in metrics:
    for i, cond in enumerate(conditions):
        summary_rows.append({
            'Metric': metric,
            'Condition': cond.capitalize(),
            'Mean': f"{means[metric][i]:.2f}",
            'SD': f"{stds[metric][i]:.2f}",
            'SE': f"{ses[metric][i]:.2f}",
            'M (SD)': f"{means[metric][i]:.2f} ({stds[metric][i]:.2f})"
        })

summary_df = pd.DataFrame(summary_rows)
summary_path = '/Users/ge93mey/Desktop/MIT_Media_Lab/Cyborg Psychology/IUI_Analysis/Results/Graphs/interaction_metrics_summary.csv'
os.makedirs(os.path.dirname(summary_path), exist_ok=True)
summary_df.to_csv(summary_path, index=False)
print(f"âœ… Summary table saved to: {summary_path}")
```

## Mediation Analysis - POST score predicition based on PRE and Interaction Scores, ie. Do PRE scores AND interaction experiences predict POST scores? POST_metric ~ PRE_metric + INT_UES + INT_Realism + INT_Trust + INT_Persuasion

```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from statsmodels.stats.multitest import multipletests
import statsmodels.api as sm  # only needed if you still use the old OLS style elsewhere

# =========================
# CONFIG
# =========================
post_metrics = ['anxious','overwhelmed','relaxed','energetic','motivated',
                'AHS','FSQ_Similarity','FSQ_Vividness','FSQ_Positivity','Future_self_continuity']

pre_predictors = ['anxious','overwhelmed','relaxed','energetic','motivated',
                  'AHS','FSQ_Similarity','FSQ_Vividness','FSQ_Positivity','Future_self_continuity']

cond = ['avatar', 'text', 'voice', 'control']  # not used directly here but may be useful
interaction_predictors = ['UES','Realism','Trust','Persuasion']

# =========================
# BUILD `wide` (your logic)
# =========================

# Ensure user_id is a column
for df in (PRE_final, POST_interaction_final, POST_final):
    if 'user_id' not in df.columns:
        df['user_id'] = df.index

# --- PRE: prefix with PRE_ ---
pre_w = PRE_final[['user_id'] + pre_predictors].copy()
pre_w = pre_w.rename(columns={c: f'PRE_{c}' for c in pre_predictors})

# --- INTERACTION: prefix with INT_ ---
inter_w = POST_interaction_final[['user_id'] + interaction_predictors].copy()
inter_w = inter_w.rename(columns={c: f'INT_{c}' for c in interaction_predictors})

# --- POST: handle potential name/case differences and prefix with POST_ ---
post_cols_present = set(POST_final.columns)
def post_name(c):
    titled = c[0].upper() + c[1:]
    if c in post_cols_present:
        return c
    if titled in post_cols_present:
        return titled
    return c  # last resort

post_map = {c: post_name(c) for c in post_metrics}
post_w = POST_final[['user_id'] + list(post_map.values())].copy()
post_w = post_w.rename(columns={post_map[c]: f'POST_{c}' for c in post_metrics})



# ---- NEW: bring condition in from PRE_final ----
cond_w = PRE_final[['user_id', 'cond']].drop_duplicates('user_id')



# --- MERGE ALL INTO `wide` ---
wide = pre_w.merge(inter_w, on='user_id', how='inner').merge(post_w, on='user_id', how='inner').merge(cond_w, on='user_id', how='inner')

# Coerce all non-id columns to numeric (invalid -> NaN)
for c in wide.columns:
    if c != 'user_id':
        wide[c] = pd.to_numeric(wide[c], errors='coerce')

print(f"wide shape: {wide.shape}")

# =========================
# REGRESSION: POST ~ PRE + INT
# =========================

# helper: p-value -> stars
def significance_stars(p):
    if p < 0.001:
        return '***'
    elif p < 0.01:
        return '**'
    elif p < 0.05:
        return '*'
    else:
        return 'n.s.'

regression_results = []
coeff_rows = []
fitted_models = {}

if wide.empty:
    print("âš ï¸ `wide` is empty after merging. No regressions will be run.")
else:
    # --- Interaction columns actually present in wide ---
    inter_cols = [f'INT_{c}' for c in interaction_predictors if f'INT_{c}' in wide.columns]
    print(f"Using interaction predictors: {inter_cols}")

    # --- PRE predictors actually present in wide ---
    pre_cols = [f'PRE_{m}' for m in pre_predictors if f'PRE_{m}' in wide.columns]

    X_cols = pre_cols + inter_cols
    print(f"Predictors ({len(X_cols)} total):", X_cols)

    if not X_cols:
        print("âš ï¸ No predictor columns found in `wide` (no PRE_ or INT_ columns). No regressions will be run.")
    else:
        # --- run regression for each POST outcome ---
        for target in post_metrics:
            y_col = f'POST_{target}'
            if y_col not in wide.columns:
                print(f"âš ï¸ Skipping {target}: missing {y_col}")
                continue

            needed = ['user_id', y_col] + X_cols
            present = [c for c in needed if c in wide.columns]
            missing = [c for c in needed if c not in wide.columns]

            if missing:
                print(f"âš ï¸ Skipping {y_col}: missing {missing}")
                continue

            sub = wide[present].copy()
            sub = sub.replace([np.inf, -np.inf], np.nan).dropna(subset=[y_col] + X_cols)

            if sub.empty:
                print(f"âš ï¸ Skipping {target}: no valid rows after cleaning.")
                continue

            # --- fit robust OLS (HC3) ---
            formula = f'{y_col} ~ ' + ' + '.join(X_cols)
            model = smf.ols(formula, data=sub).fit(cov_type='HC3')
            fitted_models[y_col] = model

            print(f"\n=== Regression: {y_col} ~ all PRE + interaction metrics ===")
            print(model.summary())

            # --- model-level info ---
            regression_results.append({
                'Dependent Variable': y_col,
                'R-squared': model.rsquared,
                'Adj R-squared': model.rsquared_adj,
                'F-statistic': model.fvalue,
                'F-test p-value': model.f_pvalue,
                'df_model': int(model.df_model),
                'df_resid': int(model.df_resid),
                'N': int(model.nobs)
            })

            # --- coefficient-level info ---
            y_sd = sub[y_col].std(ddof=1)
            X_sds = sub[X_cols].std(ddof=1)
            conf_int = model.conf_int()

            for predictor in model.params.index:
                b = model.params[predictor]
                se = model.bse[predictor]
                t = model.tvalues[predictor]
                p = model.pvalues[predictor]
                ci_low, ci_high = conf_int.loc[predictor].values

                if predictor == 'Intercept':
                    beta_std = np.nan
                    ci_beta_low = np.nan
                    ci_beta_high = np.nan
                else:
                    if predictor in X_sds.index and y_sd > 0 and X_sds[predictor] > 0:
                        scale = X_sds[predictor] / y_sd
                        beta_std = b * scale
                        ci_beta_low = ci_low * scale
                        ci_beta_high = ci_high * scale
                    else:
                        beta_std = np.nan
                        ci_beta_low = np.nan
                        ci_beta_high = np.nan

                coeff_rows.append({
                    'Dependent Variable': y_col,
                    'Predictor': predictor,
                    'B_unstd': b,
                    'SE_B': se,
                    'Beta_std': beta_std,
                    't': t,
                    'p': p,
                    'p_stars': significance_stars(p),  # raw p stars
                    'CI_B_low': ci_low,
                    'CI_B_high': ci_high,
                    'CI_Beta_low': ci_beta_low,
                    'CI_Beta_high': ci_beta_high,
                    'R-squared': model.rsquared,
                    'Adj R-squared': model.rsquared_adj,
                    'N': int(model.nobs)
                })

# =========================
# TABLES + FDR
# =========================
if not coeff_rows:
    print("\nâš ï¸ No coefficients estimated. Check that `wide` has POST_, PRE_, and INT_ columns and valid rows.")
    model_summary_df = pd.DataFrame()
    final_table = pd.DataFrame()
    significant_only = pd.DataFrame()
    significant_FDR_only = pd.DataFrame()
else:
    model_summary_df = pd.DataFrame(regression_results)

    final_table = pd.DataFrame(coeff_rows).sort_values(
        by=['Dependent Variable', 'Predictor']
    ).reset_index(drop=True)

    # --- FDR CORRECTION (Benjaminiâ€“Hochberg) ---
    mask = final_table['Predictor'] != 'Intercept'
    pvals = final_table.loc[mask, 'p'].values

    reject, pvals_corr, _, _ = multipletests(pvals, alpha=0.05, method='fdr_bh')

    final_table.loc[mask, 'p_FDR'] = pvals_corr
    final_table.loc[mask, 'FDR_significant'] = reject
    final_table.loc[mask, 'p_FDR_stars'] = [significance_stars(p) for p in pvals_corr]

    # mark intercepts
    final_table.loc[~mask, ['p_FDR', 'FDR_significant', 'p_FDR_stars']] = [np.nan, False, '']

    # convenience views
    significant_only = final_table[
        (final_table['Predictor'] != 'Intercept') & (final_table['p'] < 0.05)
    ].copy()

    significant_FDR_only = final_table[
        (final_table['Predictor'] != 'Intercept') & (final_table['FDR_significant'])
    ].copy()

# At this point you can:
# display(model_summary_df)
# display(final_table)
# display(significant_only)
# display(significant_FDR_only)
```

### FDR Correction

```python
# Exclude intercept from multiple-comparison correction
mask = final_table['Predictor'] != 'Intercept'

pvals = final_table.loc[mask, 'p'].values

reject, pvals_corr, _, _ = multipletests(
    pvals,
    alpha=0.05,
    method='fdr_bh'     # Benjaminiâ€“Hochberg
)

# Store corrected p-values and significance flags
final_table.loc[mask, 'p_FDR'] = pvals_corr
final_table.loc[mask, 'FDR_significant'] = reject

# For intercept rows, leave NA
final_table.loc[~mask, 'p_FDR'] = np.nan
final_table.loc[~mask, 'FDR_significant'] = False

# Convenience filtered views
significant_only = final_table[
    (final_table['Predictor'] != 'Intercept') & (final_table['p'] < 0.05)
].copy()

significant_FDR_only = final_table[
    (final_table['Predictor'] != 'Intercept') & (final_table['FDR_significant'] == True)
].copy()

print("\n=== SIGNIFICANT PREDICTORS AFTER FDR (q < .05, excl. intercept) ===")
try:
    display(significant_FDR_only)
except NameError:
    print(significant_FDR_only.to_string(index=False))
```

```python
import matplotlib.pyplot as plt

# forest plot for each outcome
for target in post_metrics:
    dv = f'POST_{target}'
    df_dv = final_table[
        (final_table['Dependent Variable'] == dv) &
        (final_table['Predictor'] != 'Intercept')
    ].copy()

    if df_dv.empty:
        continue

    # sort by absolute standardized beta
    df_dv = df_dv.sort_values(by='Beta_std', key=lambda s: s.abs())

    preds = df_dv['Predictor'].values
    betas = df_dv['Beta_std'].values
    ci_low = df_dv['CI_Beta_low'].values
    ci_high = df_dv['CI_Beta_high'].values

    # x positions
    y_pos = np.arange(len(preds))

    plt.figure(figsize=(6, max(4, 0.4 * len(preds))))
    # central point
    plt.scatter(betas, y_pos, marker='o')

    # horizontal CI lines
    for i in range(len(preds)):
        if np.isfinite(ci_low[i]) and np.isfinite(ci_high[i]):
            plt.plot([ci_low[i], ci_high[i]], [y_pos[i], y_pos[i]], 'k-', linewidth=1)

    plt.axvline(0, color='grey', linestyle='--', linewidth=1)

    plt.yticks(y_pos, preds, fontsize=8)
    plt.xlabel('Standardized coefficient (Î²)')
    plt.title(f'Regression coefficients for {dv}')
    plt.tight_layout()
    plt.show()
```
